```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.loadtxt('ex1data1.txt', dtype='float', delimiter=',')
```

```python
x_train=data[:,0];y_train=data[:,1]
print(type(x_train))
```

```python
print(x_train[:5]);print(y_train[:5])
```

```python
fig=plt.figure()
ax= fig.add_subplot()
ax.scatter(x_train,y_train,marker='o', color='blue')
ax.set_xlabel('City population in 10,000')
ax.set_ylabel('profit in 10,000')
ax.set_title('profit Vs population')
```

```python
def cost_function(w,b,x,y):
    #x: feature vector: ndarray of shape (m,)
    #y: output vector: ndarray of shape (m,)
    m=x.shape[0]
    cost= (1/(2*m))*np.sum(np.square(w*x+b-y))
    return cost
```

```python
def update_w_b(w,b,x,y,alpha,num_iter):
    m=x.shape[0]
    cost_history=[];
    for i in range(num_iter):
        dj_db=(1/m)*np.sum(w*x+b-y)
        dj_dw=(1/m)*np.sum(np.dot(w*x+b-y,x))
        b=b-alpha*dj_db
        w=w-alpha*dj_dw
        cost=cost_function(w,b,x,y)
        cost_history.append(cost)
    
    
    return w,b,cost_history
```

```python
def predict(w,b,x):
    y_predict=w*x+b

    return y_predict
```

```python
initial_w=0;initial_b=0;alpha=0.01;num_iter=1500;
w,b,cost_history= update_w_b(initial_w,initial_b,x_train,y_train,alpha,num_iter)
print(f"Cost found after {num_iter} iteration is {cost_history[num_iter-1]}")
print(f"W and b after {num_iter} iteration are {w} and {b}, respectively.")
#print(w,b,cost_history)
```

```python
plt.scatter(np.linspace(1.0, 1500.0, num=1500),cost_history)
```

```python
y_predicted=predict(w,b,x_train)
fig=plt.figure()
ax=fig.add_subplot()
ax.scatter(x_train,y_predicted,label='Predicted')
ax.scatter(x_train,y_train,marker='*',color='red',label='Dataset')
ax.legend()
ax.set_xlabel('City population in 10,000')
ax.set_ylabel('profit in 10,000')
ax.set_title('profit Vs population')
```
